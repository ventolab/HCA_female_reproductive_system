# 00aiii · Decoupler/PyDESeq2 pseudobulk DEG analysis, one vs. rest method (Mac_LYVE1hi, uftLAM, oLAM)

Marker identification with differential gene expression (DEG) analysis.

**Pipeline:** based on the [**Pseudobulk Enrichment Analysis** tutorial](https://decoupler.readthedocs.io/en/latest/notebooks/scell/rna_psbk.html) by `decoupler`.
1. Load annotated .h5ad object containing raw counts.
2. Subset AnnData to groups (typically cell types) of interest in order to perform targeted DEG analysis.
3. Pseudobulk the data by donor x group.
4. Filter out low-quality pseudobulk samples and lowly expressed genes.
5. Run DESeq2 to identify differentially expressed genes using a **one vs. rest method.**
6. Plot top DEGs on a volcano plot.
7. Save all upregulated and downregulated DEGs.

## Configuration

```python
import os, sys

# ── Paths ─────────────────────────────────────────────────────────────────────
OBJECT          = "acrosslifespan_noHormones_immune"
MAIN_DIR        = "/nfs/team292/projects/PanTissue/"
OUTPUT_DIR      = os.path.join(MAIN_DIR, "results/temp/02_annotation/immune/acrosslifespan/noHormones/")
FIG_DIR         = os.path.join(OUTPUT_DIR, "figures/")
ANNOTATED_H5AD  = os.path.join(OUTPUT_DIR, f"annotated_{OBJECT}.h5ad")

os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(FIG_DIR, exist_ok=True)

# ── Adjustable parameters ───────────────────────────────────────────────——————
DONOR_COL         = "Donor_id"
ANNOT_COL         = "HCA_celltype_label"
GROUPS            = [
    "Immune_Mac_LYVE1hi",
    "Immune_Mac_FCGBPhi", # aka uftLAM - this is a previous name
    "Immune_Mac_DHRS9hi"  # aka oLAM - this is a previous name
]

MIN_CELLS  = 10
MIN_COUNTS = 1000
```

## Imports

```python
import warnings
warnings.filterwarnings("ignore")

import anndata as ad
import scanpy as sc
import pandas as pd
import numpy as np

import decoupler as dc
from pydeseq2.dds import DeseqDataSet, DefaultInference
from pydeseq2.ds import DeseqStats

import matplotlib.pyplot as plt
```

# 1. Load annotated object

```python
adata = ad.read_h5ad(ANNOTATED_H5AD)
adata
```

# 2. Subset to cells of interest

```python
print(f"Subsetting to only include {', '.join(GROUPS)}...")
adata_sub = adata[adata.obs[ANNOT_COL].isin(GROUPS)].copy()

print(f">> {adata_sub.n_obs} cells kept\n")
adata_sub.obs[ANNOT_COL].value_counts()
```

```python
adata_sub.obs.columns
```

# 3. Pseudobulk counts

**Aggregate raw counts using `decoupler`:**

`pdata` is an AnnData object, where:
* Each row (`obs`) is one pseudobulk sample (donor x cell type)
* Each column (`var`) is a gene
* `pdata.obs` contains donor, cell type, and QC columns
    * `psbulk_cells`: number of cells aggregated
    * `psbulk_counts`: total counts in that pseudobulk sample

Note that `obs` columns are only carried over if it has the **same value** across the pseudobulk level (i.e., donor).

```python
# —— Aggregate counts per donor × cell type using decoupler ————————————
pdata = dc.pp.pseudobulk(
    adata      = adata_sub,
    sample_col = DONOR_COL,   # column in adata.obs with donor/sample IDs
    groups_col = ANNOT_COL,   # column in adata.obs with cell type labels
    mode       = "sum"
)
pdata
```

The following plot shows each pseudobulk sample as a point, with number of cells on one axis and total counts on the other. Samples in the lower-left corner are likely low quality, and can be excluded from the analysis using `dc.pp.filter_samples()`.

Filter thresholds are always dataset-dependent and arbitrary, but a common guideline is to retain samples with at least **10 cells** and **1000 total counts.**

```python
# —— Visualise QC metrics before filtering —————————————————————————————
QC_METRICS = [DONOR_COL, ANNOT_COL]

for METRIC in QC_METRICS:
    # —— Conditionally suppress legend if too many group levels ————————
    n_levels = pdata.obs[METRIC].nunique()
    legend_setting = "auto" if n_levels <= 10 else False
    
    dc.pl.filter_samples(
        adata          = pdata,
        groupby        = [METRIC],
        min_cells      = MIN_CELLS,
        min_counts     = MIN_COUNTS,
        figsize        = (6,3),
        kw_scatterplot = {"legend": legend_setting}
        # kw_scatterplot passes extra keyword arguments directly to
        # seaborn.scatterplot, which draws the points in the QC plot
    )
```

```python
# —— Apply the filter ——————————————————————————————————————————————————
pre_filt_cells = pdata.n_obs
print(f"Removing samples with at least {MIN_CELLS} cells and {MIN_COUNTS} counts...")
dc.pp.filter_samples(
    pdata,
    min_cells  = MIN_CELLS,
    min_counts = MIN_COUNTS
)
post_filt_cells = pdata.n_obs
print(f">> {pre_filt_cells - post_filt_cells} pseudobulk samples removed.\n")
pdata
```

# 4. Filter genes

Lowly expressed genes will also be filtered prior to downstream analysis. This step should be performed at the group (i.e., cell type) level, as different cell types may express distinct sets of genes. The following step uses the default parameters as specified in the `decoupler` Pseudobulk Enrichment Analysis tutorial.

The following function stores the filtered pseudobulk AnnDatas for each subset in a dictionary, keyed by subset name, so it is ready to pass directly into DESeq2 in the next step without recomputing.

```python
# —— Filter genes per cell type and store results ——————————————————————
filtered_pdata = {}

for subset in GROUPS:

    print(f"\nFiltering genes for: {subset}...")

    # —— Subset pseudobulk data to this cell type only —————————————————
    pdata_sub = pdata[pdata.obs[ANNOT_COL] == subset].copy()

    # —— Filter lowly expressed genes using decoupler ——————————————————
    dc.pp.filter_by_expr(
        adata           = pdata_sub,
        group           = ANNOT_COL,
        min_count       = 10,
        min_total_count = 15,
        large_n         = 10,
        min_prop        = 0.7
    )

    dc.pp.filter_by_prop(
        adata     = pdata_sub,
        min_prop  = 0.1,
        min_smpls = 2
    )

    print(f">> Genes remaining after filtering: {pdata_sub.shape[1]}")

    filtered_pdata[subset] = pdata_sub
```

# 5. Run DESeq2

In this notebook, we run DEG analysis via a **one vs. rest** method, analogous to how other DEG methods (such as those in Scanpy) identify *de novo* markers for cell types. This analysis can thus be used as a complement to TF-IDF marker discovery.

**Run one vs. rest for each subtype in GROUPS:** each iteration runs the OvR comparison for one group and stores both the full and filtered results in dictionaries, keyed by the subtype name (e.g. `all_results["subset1"], all_sig_results["subset2"]`)

```python
# —— Define a reusable DEG filter function —————————————————————————————
def filter_degs(
    results_df,
    padj_thresh:  float=0.05,
    lfc_thresh:   float=1.0
):
    """
    Filters a PyDESeq2 results DataFrame to significant DEGs.

    Parameters
    ----------
    results_df : pd.DataFrame
        Output from DeseqStats.results_df
    padj_thresh : float
        Adjusted p-value threshold. Default 0.05
    lfc_thresh : float
        Absolute log2 fold change threshold. Default 1.0 (i.e. 2-fold)
    """
    mask = (
        (results_df["padj"] < padj_thresh) &
        (results_df["log2FoldChange"].abs() > lfc_thresh)
    )
    return results_df[mask].sort_values("padj")
```

```python
# —— Run one-vs-rest DESeq2 for each cell type —————————————————————————
all_results     = {}
all_sig_results = {}

for subset in GROUPS:

    print(f"\nRunning DESeq2 for: {subset}")

    # —— Build binary one-vs-rest label on the FULL pdata —————————————
    pdata.obs["one_vs_rest"] = np.where(
        pdata.obs[ANNOT_COL] == subset,
        "target",
        "rest"
    )

    # —— Retrieve the gene list from the filtered subset in Step 6a ————
    genes_to_keep = filtered_pdata[subset].var_names

    # —— Subset pdata to the filtered genes only ———————————————————————
    pdata_sub = pdata[:, genes_to_keep].copy()

    # —— Extract counts and metadata as DataFrames for PyDESeq2 ————————
    counts_df = pd.DataFrame(
        pdata_sub.X,
        index   = pdata_sub.obs_names.astype(str),
        columns = pdata_sub.var_names
    )
    metadata_df = pdata_sub.obs[["one_vs_rest"]].copy()
    metadata_df.index = metadata_df.index.astype(str)

    # —— Initialise and fit DESeq2 —————————————————————————————————————
    inference = DefaultInference(n_cpus=8)

    dds = DeseqDataSet(
        counts      = counts_df,
        metadata    = metadata_df,
        design      = "~one_vs_rest",
        refit_cooks = True,
        inference   = inference
    )

    dds.deseq2()

    # —— Extract contrast and run Wald test ————————————————————————————
    stat_res = DeseqStats(
        dds,
        contrast  = ["one_vs_rest", "target", "rest"],
        inference = inference
    )
    stat_res.summary()

    results               = stat_res.results_df
    results["comparison"] = f"{subset}_vs_rest"

    # —— Filter to significant DEGs ————————————————————————————————————
    sig_results = filter_degs(
        results,
        padj_thresh = 0.05,
        lfc_thresh  = 1.0
    )

    all_results[subset]     = results
    all_sig_results[subset] = sig_results

# —— Combine all results ———————————————————————————————————————————————
all_results_df     = pd.concat(all_results.values(),     axis=0)
all_sig_results_df = pd.concat(all_sig_results.values(), axis=0)

print(f"\nTotal significant DEGs: {len(all_sig_results_df)}")
print(all_sig_results_df.groupby("comparison").size())
```

# 6. Plot top DEGs

```python
# —— Plot a volcano for each comparison ———————————————————————————————
for subset, results in all_results.items():
    
    fig = dc.pl.volcano(
        data       = results,
        x          = "log2FoldChange",
        y          = "pvalue",
        top        = 10,
        figsize    = (5, 4),
        return_fig = True
    )

    fig.axes[0].set_title(
        f"{subset} vs rest",
        fontsize = 12,
        pad      = 10
    )

    fig.savefig(f"{FIG_DIR}volcano_01_OvR_{subset}_{OBJECT}.png", bbox_inches="tight")
    plt.show()
```

# 7. Save all DEGs

```python
# —— Build summary DEG table across all subsets ————————————————————————
deg_summary = []

for subset, sig_results in all_sig_results.items():

    subset_df = pd.DataFrame({
        "gene_name": sig_results.index,
        ANNOT_COL:   subset,
        "log2FC":    sig_results["log2FoldChange"].values,
        "padj":      sig_results["padj"].values
    })

    deg_summary.append(subset_df)

# —— Concatenate all subsets into one DataFrame ————————————————————————
deg_summary_df = pd.concat(deg_summary, axis=0).reset_index(drop=True)

# —— Split into upregulated and downregulated —————————————————————————
deg_up   = deg_summary_df[deg_summary_df["log2FC"] > 0].reset_index(drop=True)
deg_down = deg_summary_df[deg_summary_df["log2FC"] < 0].reset_index(drop=True)

# —— Save to CSV ———————————————————————————————————————————————————————
deg_up.to_csv(
    f"{OUTPUT_DIR}DESeq2_upDEGs_OvR_TRMs_{OBJECT}.csv",
    index = False
)

deg_down.to_csv(
    f"{OUTPUT_DIR}DESeq2_downDEGs_OvR_TRMs_{OBJECT}.csv",
    index = False
)

# —— Print summary ————————————————————————————————————————————————————
print(f"Upregulated DEGs:   {len(deg_up)}")
print(deg_up.groupby(ANNOT_COL).size())

print(f"\nDownregulated DEGs: {len(deg_down)}")
print(deg_down.groupby(ANNOT_COL).size())
```
