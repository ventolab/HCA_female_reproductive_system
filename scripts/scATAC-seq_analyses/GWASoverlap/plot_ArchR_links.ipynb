```r
plotPeakCallSummary <- function(
	ArchRProj = NULL, 
	pal = NULL
	){

  peakDF <- S4Vectors::metadata(ArchRProj@peakSet)$PeakCallSummary
  
  if(is.null(peakDF)){
    stop("Error no Peak Call Summary available are you sure these peaks were called with CreateReproduciblePeakSet?")
  }

  if(is.null(pal)){
    pal <- paletteDiscrete(values=peakDF$Var1)
  }
  pal <- c("Distal" = "#60BA64", "Exonic" = "#73C6FF", "Intronic" = "#620FA3", "Promoter" = "#FFC554", pal) 
  pal <- pal[!duplicated(names(pal))]

  lengthMax <- split(peakDF$Freq, peakDF$Group) %>% lapply(sum) %>% unlist %>% max

  colnames(peakDF)[colnames(peakDF)=="Var1"] <- "PeakAnno"

  p <- ggplot(peakDF, aes(x=Group, y=Freq, fill=PeakAnno)) + 
    geom_bar(stat = "identity") + 
    theme_ArchR(xText90 = TRUE) +
    ylab("Number of Peaks (x10^3)") +
    xlab("") + 
    theme(legend.position = "bottom", 
      legend.key = element_rect(size = 2), 
      legend.box.background = element_rect(color = NA)
    ) +
    scale_fill_manual(values=pal) +
    scale_y_continuous(
      breaks = seq(0, lengthMax * 2,50), 
      limits = c(0, lengthMax * 1.1), 
      expand = c(0,0)
      )

  attr(p, "ratioYX") <- 0.5

  return(p)

}
```

```r
plot_track=function(gene,peak,groupby,geneAnno = geneAnnotation,remove_other_genes=TRUE,extend_peak=0,group_order=NULL) {
	start(peak)=start(peak)-extend_peak
	end(peak)=end(peak)+extend_peak
	if (remove_other_genes){
		genes_gr <- geneAnno$genes
		#exons_gr <- geneAnno$exons
		genes_gr$symbol <- ifelse(genes_gr$symbol == gene, genes_gr$symbol, "")
		#exons_gr$symbol <- ifelse(exons_gr$symbol == gene, exons_gr$symbol, "")
		geneAnno$genes <- genes_gr
		#geneAnno$exons <- exons_gr
	}

	gene_idx <- which(genes$name ==gene)
	p2g_filtered <- p2g[p2g$idxRNA %in% gene_idx, ]

	peak_pos <- start(peaks[p2g_filtered$idxATAC])
	gene_pos <- start(genes[p2g_filtered$idxRNA])

	loops_filtered <- GRanges(
  		seqnames = seqnames(peaks[p2g_filtered$idxATAC]),
  		ranges = IRanges(start = pmin(peak_pos, gene_pos),end   = pmax(peak_pos, gene_pos)),
		value = p2g_filtered$Correlation, FDR   = p2g_filtered$FDR)
	message(paste0("Loops: ",length(loops_filtered)))

	loops_filtered$value <- -log10(loops_filtered$FDR + 1e-100)  # small pseudocount to avoid -log10(0)

	#define window range
	# Span between the highlight peak and the gene TSS only
  	anchor1 <- min(start(peak),start(genes[gene_idx[1]]))-50000
  	anchor2 <- max(end(peak),end(genes[gene_idx[1]]))+50000
	region_gr <- GRanges(seqnames = seqnames(peak),ranges   = IRanges(start = anchor1, end = anchor2))

	if (is.null(group_order)){
		useGroups=unique(na.omit(getCellColData(proj, groupby, drop = TRUE)))
	} else {
		useGroups=group_order
	}
	p <- plotBrowserTrack(ArchRProj = proj, groupBy = groupby, 
	useGroups = useGroups,
    geneSymbol = gene, 
	highlight=peak,
	region     = region_gr,upstream   = 0,downstream = 0,
    loops = loops_filtered, geneAnnotation=geneAnno)
	return(p)
}
```
